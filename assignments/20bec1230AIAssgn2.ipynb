```python
#task1---preprocessing
import numpy as np
import pandas as pd
```

task-1  ------> data preprocessing

```python
#reading data
dfrm = pd.read_csv('drugsAssgn2.csv')
print(dfrm.head(),'\n')
```

```python
#feature extraction
x = dfrm.iloc[:,0:5]
print(x.head(),'\n')
y = dfrm.iloc[:,5]
print(x.head(),'\n')
```

```python
#traintest split
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte = train_test_split(x,y,test_size=0.2,random_state=10)
#encoding
xtr = pd.get_dummies(xtr, columns=['Sex','BP','Cholesterol'])
xte = pd.get_dummies(xte, columns=['Sex','BP','Cholesterol'])

num_classes = len(np.unique(y))
ytr = pd.get_dummies(ytr).values
yte = pd.get_dummies(yte).values

print('after encoding input values are','\n')
xtr.head()
```

```python
print('target variable encoding : ','\n')
print(ytr[0:20])
```

TASK-2------MODEL FIT

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(10, activation='relu', input_dim=xtr.shape[1]))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(xtr, ytr, epochs=1000, batch_size=32, verbose=1)
```

TASK-3  -----  EVALUATION

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(xte, yte, verbose=0)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```

```python
dfrand = pd.DataFrame({'age':[23],'Sex':['F'],'BP':['HIGH'],'Cholesterol':['HIGH'],'Na_to_K':[25.355]})
#we can change values in the above to obtain required random input
dfrandE = pd.get_dummies(dfrand, columns=['Sex','BP','Cholesterol'])
dfrandE = dfrandE.reindex(columns=xtr.columns, fill_value=0)
xPred = model.predict(dfrandE)
labels=['DrugY','drugA','drugB','drugC','drugX']
y_decoded = [labels[np.argmax(row)] for row in xPred]
print('predicted drug:',y_decoded,'----->','which is correct according to the given raw data')

```
