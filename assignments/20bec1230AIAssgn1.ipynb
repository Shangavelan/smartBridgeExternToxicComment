```python
import numpy as np
import pandas as pd
#question 1(random values dataframe):
dict = {'col1':np.random.rand(4),'col2':np.random.rand(4),'col3':np.random.rand(4),'col4':np.random.rand(4)}
dfrm = pd.DataFrame(dict)
dfrm
```

```python
#question 2
dfrm=dfrm.rename(columns={'col1':'Random Value 1','col2':'Random Value 2','col3':'Random Value 3','col4':'Random Value 4'})
dfrm
```

```python
#question 3
dfrm.describe()
```

```python
#question 4
nulvalsAll = dfrm.isnull()
nulssum=nulvalsAll.sum()
print(nulvalsAll,"\n")
print(nulssum,'\n')
print('type of col1 : ',dfrm['Random Value 1'].dtype)
print('type of col2 : ',dfrm['Random Value 2'].dtype)
print('type of col3 : ',dfrm['Random Value 3'].dtype)
print('type of col4 : ',dfrm['Random Value 4'].dtype)
```
